```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
```

```python
te_data=te.fit(data).transform(data)
```

```python
te.columns_
```

```python
df2=pd.DataFrame(te_data,columns=te.columns_)
```

```python
df2
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
import numpy as np
```

```python
df3=apriori(df2,min_support=0.15,use_colnames=True)
```

```python
df3
```

```python
df3.sort_values(by='support',ascending=False)
```

```python
df4=association_rules(df3,metric='confidence',min_threshold=0.6)
```

```python
df4
```

```python
filter=df4[(df4['support']>=0.15) & (df4['confidence']>=0.75)]
```

```python
filter
```
